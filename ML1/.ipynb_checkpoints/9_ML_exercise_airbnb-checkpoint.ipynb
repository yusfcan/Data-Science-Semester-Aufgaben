## Teil 1: Setup und Daten laden

```python
# Bibliotheken importieren
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score

# Visualisierung
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')

print("✓ Setup komplett")
```

```python
# Daten laden (bereinigte Version aus Woche 7)
df = pd.read_csv('airbnb_berlin_cleaned.csv')

print(f"Datensatz: {df.shape[0]} Zeilen, {df.shape[1]} Spalten")
print(f"\nErste 5 Zeilen:")
display(df.head())

print(f"\nVerfügbare Spalten:")
print(df.columns.tolist())
```

## Teil 2: Schnelle EDA

Kurzer Check der wichtigsten Variablen für ML

```python
# Numerische Spalten
print("Numerische Features:")
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(numeric_cols)

# Beschreibende Statistik
print("\nStatistik der wichtigsten Features:")
display(df[['bedrooms', 'accommodates', 'price_num', 'number_of_reviews']].describe())
```

```python
# Room Type Verteilung
print("Room Type Verteilung:")
print(df['room_type'].value_counts())

plt.figure(figsize=(10, 5))
df['room_type'].value_counts().plot(kind='bar', color='steelblue')
plt.title('Anzahl Listings pro Room Type')
plt.xlabel('Room Type')
plt.ylabel('Anzahl')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Teil 3: Feature Selection für Preis-Vorhersage

Welche Features korrelieren mit dem Preis?

```python
# TODO: Berechne Korrelation mit price_num
# DEIN CODE HIER

# Nur numerische Spalten für Korrelation
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Korrelation mit price_num
if 'price_num' in numeric_df.columns:
    correlation = numeric_df.corr()['price_num'].sort_values(ascending=False)
    print("Korrelation mit Preis:")
    print(correlation)
    
    # Visualisierung
    plt.figure(figsize=(10, 6))
    correlation_plot = correlation[1:]  # Ohne price_num selbst
    colors = ['green' if x > 0 else 'red' for x in correlation_plot.values]
    plt.barh(range(len(correlation_plot)), correlation_plot.values, color=colors)
    plt.yticks(range(len(correlation_plot)), correlation_plot.index)
    plt.xlabel('Korrelation mit Preis')
    plt.title('Welche Features beeinflussen den Preis?')
    plt.axvline(x=0, color='black', linestyle='--', linewidth=0.8)
    plt.tight_layout()
    plt.show()
    
    print(f"\n Stärkste Korrelation: {correlation.index[1]} ({correlation.values[1]:.3f})")
```

## Teil 4: Problem 1 - REGRESSION (Preis vorhersagen)

**Fragestellung:** Wie viel kostet eine Airbnb-Unterkunft mit X Schlafzimmern für Y Gäste?

**Typ:** Regression (kontinuierlicher Wert)

**Features:** bedrooms, accommodates

**Target:** price_num

```python
# Daten vorbereiten für Regression
df_regression = df[['bedrooms', 'accommodates', 'price_num']].dropna()

# Features und Target
X_reg = df_regression[['bedrooms', 'accommodates']]
y_reg = df_regression['price_num']

print(f"Regression Dataset: {X_reg.shape[0]} Samples")
print(f"Features: {X_reg.columns.tolist()}")
print(f"Target: price_num (Durchschnitt: {y_reg.mean():.2f}€)")
```

```python
# TODO: Train/Test Split für Regression (80/20)
# DEIN CODE HIER

X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(
    X_reg, y_reg,
    test_size=0.2,
    random_state=42
)

print(f"Training Set: {len(X_reg_train)} Samples")
print(f"Test Set: {len(X_reg_test)} Samples")
```

### Regression: Verschiedene Modelle vergleichen

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

# TODO: Definiere verschiedene Regression-Modelle
# DEIN CODE HIER

regression_models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(max_depth=5, random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=50, max_depth=5, random_state=42),
    'K-Nearest Neighbors': KNeighborsRegressor(n_neighbors=5)
}

print(f"✓ {len(regression_models)} Modelle definiert")
```

```python
# Alle Modelle trainieren und evaluieren
regression_results = []

for name, model in regression_models.items():
    # Training
    model.fit(X_reg_train, y_reg_train)
    
    # Vorhersagen
    y_pred_train = model.predict(X_reg_train)
    y_pred_test = model.predict(X_reg_test)
    
    # Metriken
    train_r2 = r2_score(y_reg_train, y_pred_train)
    test_r2 = r2_score(y_reg_test, y_pred_test)
    test_rmse = np.sqrt(mean_squared_error(y_reg_test, y_pred_test))
    
    regression_results.append({
        'Model': name,
        'Train R²': train_r2,
        'Test R²': test_r2,
        'Test RMSE': test_rmse
    })
    
    print(f" {name:25s} - Test R²: {test_r2:.3f} | RMSE: {test_rmse:.2f}€")

# Ergebnisse als DataFrame
reg_results_df = pd.DataFrame(regression_results).sort_values('Test R²', ascending=False)
print("\n" + "="*70)
print("REGRESSION ERGEBNISSE:")
print("="*70)
display(reg_results_df)

print("\n R² Score: Wie gut erklärt das Modell die Varianz? (1.0 = perfekt)")
print(" RMSE: Durchschnittlicher Fehler in Euro")
```

```python
# Visualisierung: Bestes vs. Schlechtestes Modell
best_reg_model_name = reg_results_df.iloc[0]['Model']
worst_reg_model_name = reg_results_df.iloc[-1]['Model']

best_reg_model = regression_models[best_reg_model_name]
worst_reg_model = regression_models[worst_reg_model_name]

# Vorhersagen beider Modelle
y_pred_best = best_reg_model.predict(X_reg_test)
y_pred_worst = worst_reg_model.predict(X_reg_test)

# WICHTIG: Filtern für bessere Visualisierung (Outliers entfernen)
mask = (y_reg_test < 500) & (y_pred_best < 500) & (y_pred_worst < 500)
y_test_plot = y_reg_test[mask]
y_pred_best_plot = y_pred_best[mask]
y_pred_worst_plot = y_pred_worst[mask]

print(f"\n Visualisierung: {len(y_test_plot)} von {len(y_reg_test)} Datenpunkten")
print(f"   (Outliers > 500€ ausgeblendet für bessere Darstellung)")

# Plot erstellen
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Bestes Modell
ax[0].scatter(y_test_plot, y_pred_best_plot, alpha=0.6, s=30, color='green')
ax[0].plot([0, 500], [0, 500], 'r--', linewidth=2, label='Perfekte Vorhersage')
ax[0].set_xlabel('Tatsächlicher Preis (€)', fontsize=11)
ax[0].set_ylabel('Vorhergesagter Preis (€)', fontsize=11)
ax[0].set_title(f'✅ BESTES: {best_reg_model_name}\nR² = {reg_results_df.iloc[0]["Test R²"]:.3f}', fontsize=12)
ax[0].legend()
ax[0].grid(True, alpha=0.3)
ax[0].set_xlim(0, 500)
ax[0].set_ylim(0, 500)

# Plot 2: Schlechtestes Modell
ax[1].scatter(y_test_plot, y_pred_worst_plot, alpha=0.6, s=30, color='red')
ax[1].plot([0, 500], [0, 500], 'r--', linewidth=2, label='Perfekte Vorhersage')
ax[1].set_xlabel('Tatsächlicher Preis (€)', fontsize=11)
ax[1].set_ylabel('Vorhergesagter Preis (€)', fontsize=11)
ax[1].set_title(f'❌ SCHLECHTESTES: {worst_reg_model_name}\nR² = {reg_results_df.iloc[-1]["Test R²"]:.3f}', fontsize=12)
ax[1].legend()
ax[1].grid(True, alpha=0.3)
ax[1].set_xlim(0, 500)
ax[1].set_ylim(0, 500)

plt.tight_layout()
plt.show()

print(f"\n Bestes Modell: {best_reg_model_name} Test R²: {reg_results_df.iloc[0]['Test R²']:.3f}")

print(f"\n❌ Schlechtestes Modell: {worst_reg_model_name} Test R²: {reg_results_df.iloc[-1]['Test R²']:.3f}")

print(f"\n Vergleich:")
print(f"   - GRÜN (links): Punkte liegen näher an der roten Linie → bessere Vorhersagen")
print(f"   - ROT (rechts): Punkte streuen mehr → schlechtere Vorhersagen")
print(f"   - Der direkte Vergleich zeigt: Modellwahl macht einen großen Unterschied!")
```

## Teil 5: Problem 2 - CLASSIFICATION (Room Type vorhersagen)

**Fragestellung:** Ist eine Unterkunft mit X Schlafzimmern für Y Gäste eher ein ganzes Apartment, privates Zimmer oder Shared Room?

**Typ:** Classification (Kategorien)

**Features:** bedrooms, accommodates, number_of_reviews

**Target:** room_type

```python
# Daten vorbereiten für Classification
df_class = df[['bedrooms', 'accommodates', 'number_of_reviews', 'room_type']].dropna()

# Features und Target
X_class = df_class[['bedrooms', 'accommodates', 'number_of_reviews']]
y_class = df_class['room_type']

print(f"Classification Dataset: {X_class.shape[0]} Samples")
print(f"Features: {X_class.columns.tolist()}")
print(f"\nTarget Verteilung:")
print(y_class.value_counts())
```

```python
# TODO: Train/Test Split für Classification (80/20, stratified)
# DEIN CODE HIER

X_class_train, X_class_test, y_class_train, y_class_test = train_test_split(
    X_class, y_class,
    test_size=0.2,
    random_state=42,
    stratify=y_class  # Erhält Verteilung
)

print(f"Training Set: {len(X_class_train)} Samples")
print(f"Test Set: {len(X_class_test)} Samples")
print(f"\nTarget Verteilung (Train):")
print(y_class_train.value_counts(normalize=True))
```

### Classification: Verschiedene Modelle vergleichen

```python
from sklearn.neighbors import NearestCentroid, KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# TODO: Definiere verschiedene Classification-Modelle
# DEIN CODE HIER

classification_models = {
    'Nearest Centroid': NearestCentroid(),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Decision Tree': DecisionTreeClassifier(max_depth=5, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=50, max_depth=5, random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42)
}

print(f"✓ {len(classification_models)} Modelle definiert")
```

```python
# Alle Modelle trainieren und evaluieren
classification_results = []

for name, model in classification_models.items():
    # Training
    model.fit(X_class_train, y_class_train)
    
    # Accuracy
    train_acc = model.score(X_class_train, y_class_train)
    test_acc = model.score(X_class_test, y_class_test)
    
    classification_results.append({
        'Model': name,
        'Train Accuracy': train_acc,
        'Test Accuracy': test_acc,
        'Overfitting Gap': train_acc - test_acc
    })
    
    print(f" {name:25s} - Train: {train_acc:.2%} | Test: {test_acc:.2%}")

# Ergebnisse als DataFrame
class_results_df = pd.DataFrame(classification_results).sort_values('Test Accuracy', ascending=False)
print("\n" + "-"*70)
print("CLASSIFICATION ERGEBNISSE:")
display(class_results_df.style.format({
    'Train Accuracy': '{:.2%}',
    'Test Accuracy': '{:.2%}',
    'Overfitting Gap': '{:.2%}'
}))
```

```python
# Visualisierung der Ergebnisse
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Test Accuracy
sorted_results = class_results_df.sort_values('Test Accuracy')
colors = ['lightcoral' if x < 0.70 else 'lightgreen' for x in sorted_results['Test Accuracy']]
ax[0].barh(range(len(sorted_results)), sorted_results['Test Accuracy'], color=colors)
ax[0].set_yticks(range(len(sorted_results)))
ax[0].set_yticklabels(sorted_results['Model'])
ax[0].set_xlabel('Test Accuracy')
ax[0].set_title('Modellvergleich: Room Type Classification')
ax[0].axvline(x=0.70, color='red', linestyle='--', alpha=0.5)

# Plot 2: Train vs Test
x_pos = np.arange(len(class_results_df))
width = 0.35
ax[1].bar(x_pos - width/2, class_results_df['Train Accuracy'], width, label='Train', alpha=0.8)
ax[1].bar(x_pos + width/2, class_results_df['Test Accuracy'], width, label='Test', alpha=0.8)
ax[1].set_xlabel('Model')
ax[1].set_ylabel('Accuracy')
ax[1].set_title('Train vs. Test Accuracy')
ax[1].set_xticks(x_pos)
ax[1].set_xticklabels(class_results_df['Model'], rotation=45, ha='right')
ax[1].legend()

plt.tight_layout()
plt.show()

print(f"\n Bestes Classification Modell: {class_results_df.iloc[0]['Model']} mit Test Accuracy: {class_results_df.iloc[0]['Test Accuracy']:.2%}")
```
